# Datasets

```python
import numpy as np
import matplotlib.pyplot as make_grid
import sys
sys.path.insert(0, "../")

from datasets import get_dataset
from torch.utils.data import DataLoader
from torchvision.utils import make_grid
import matplotlib.pyplot as plt

# Sample data
mnist = DataLoader(get_dataset('mnist'), batch_size=36, shuffle=True)
mnist_x, mnist_y = next(iter(mnist))
mnist_grid = np.transpose(make_grid(mnist_x, nrow=6, normalize=True).numpy(), (1,2,0))

cifar = DataLoader(get_dataset('cifar10'), batch_size=25, shuffle=True)
cifar_x, cifar_y = next(iter(cifar))
cifar_grid = np.transpose(make_grid(cifar_x, nrow=5, normalize=True).numpy(), (1,2,0))

svhn = DataLoader(get_dataset('svhn'), batch_size=25, shuffle=True)
svhn_x, svhn_y = next(iter(svhn))
svhn_grid = np.transpose(make_grid(svhn_x, nrow=5, normalize=True).numpy(), (1,2,0))
```

```python
%matplotlib inline

fig, ax = plt.subplots(1, 3, figsize=(15, 7))

ax[0].imshow(mnist_grid)
ax[0].axis('off')
ax[0].set_title('MNIST', fontsize=16)
ax[1].imshow(cifar_grid)
ax[1].axis('off')
ax[1].set_title('CIFAR 10', fontsize=16)
ax[2].imshow(svhn_grid)
ax[2].axis('off')
ax[2].set_title('SVHN', fontsize=16)
plt.savefig('figures/dataset_samples.png', bbox_inches='tight')
plt.show()
```
